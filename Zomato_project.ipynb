# Zomato Data Analysis Project

# Step 1 - Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 - create the data frame

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
dataframe.head()
```

# Q1 what type of restaurant do the majority of customers order from?

```python
sns.countplot(x = dataframe['listed_in(type)'])
plt.xlabel('Type of Restaurant')
```

# conclusion - majority of the restaurant falls in dining category

# Q2 How many votes has each type of restaurant received from customers?

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : grouped_data})
plt.plot(result, c="green", marker ="o")
plt.xlabel("Type of restaurant", c="red", size =20)
plt.ylabel("Votes", c="red", size =20)
```

# conclusion - dining restaurant has receive more votes

# Q3 What are ratings that the majority of customers has received?

```python
plt.hist(dataframe['rate'],bins = 5)
plt.title("Ratings distribution")
plt.show()
```

# conclusion - The majority restaurants recieved ratings form 3.5 to 4

# Average order spending by couples

# Q4 Zomato has observed that most couples order most of their foods from online. What is there average spending on each other?

```python
sns.countplot(x = dataframe['approx_cost(for two people)'])
plt.xlabel('couple_data')
```

# conclusion - The majority of couples prefer restaurants with an approximate with an 300 Rs

# which mode receives maximum rating

# Q5 Which mode (online or offline) receives more rating?

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

# conclusion - offline order received lower rating in camparison to online order

# Q6 Which type of restaurant receive more offline type of order so that zomato can provide those customer with some good offers?

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu", fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# conclusion - Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. this suggest that clients prefer orders in person at restaurant , but prefer online ordering at cafes.
